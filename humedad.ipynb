```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
"""
NVCBP8A: 8. ¿La vivienda presenta:1. Humedades en el techo o en las paredes?
"""
features = ['DIRECTORIO', 'NVCBP8A']
```

```python
no_asma_2017_df = pd.read_csv('./data/data_sin_asma/Datos_proyecto_II_BI_2017_sin_asma.csv', encoding="utf-8", sep = ',')
```

```python
asma_2017_df = pd.read_csv('./data/data_asma/Datos_proyecto_II_BI_2017.csv', encoding="utf-8", sep = ',')
```

```python
no_asma_2021_df = pd.read_csv('./data/data_sin_asma/Datos_proyecto_II_BI_2021_sin_asma.csv', sep = ',', encoding='latin1')
```

```python
asma_2021_df = pd.read_csv('./data/data_asma/Datos_proyecto_II_BI_2021.csv', sep = ',', encoding='latin1')
```

```python
provisional_2017 = pd.concat([asma_2017_df[features], no_asma_2017_df[features]], ignore_index=True)
```

```python
provisional_2021 = pd.concat([asma_2021_df[features], no_asma_2021_df[features]], ignore_index=True)
```

```python
unificado = pd.concat([provisional_2017[features], provisional_2021[features]], ignore_index=True)
```

```python
(unificado[features].isnull().sum() / unificado[features].shape[0]).sort_values(ascending = False)
```

```python
# Renombrar las columnas
nuevos_nombres = {'DIRECTORIO': 'id_vivienda', 
         'NVCBP8A': 'humedad' }
unificado = unificado.rename(columns=nuevos_nombres)
```

```python
unificado.head(10)
```

```python
humedades = []

for value in unificado['humedad']:
    if value not in humedades:
        humedades.append(value)

humedades = sorted(humedades)
print(humedades)
```

```python
#Dimensión edad
df_humedad = pd.DataFrame({'humedad': humedades})

df_humedad['id_humedad'] = range(1, len(df_humedad) + 1)
df_humedad.head(10)
```

```python
unificado = pd.merge(unificado, df_humedad[['humedad', 'id_humedad']], on='humedad', how='left')
```

```python
unificado.sample(10)
```

```python
unificado = unificado.drop('humedad', axis=1)
print(unificado)
```

```python
unificado = unificado.drop_duplicates(subset='id_vivienda')
```

```python
unificado = unificado.drop('id_vivienda', axis=1)
```
